#### Question 12: Write a short program that prints the numbers 1 to 10 using a for loop. Then write an equivalent program that prints the numbers 1 to 10 using a while loop.
#### Answer : 

```python
# for loop 
for i in range(0,11):
    print(i)
```

```python
# while loop 
i=0 
while (i<=10):
    print(i)
    i+=1
```
